```python
dataFinal = dataFinal[dataFinal['Last_accr_txn_dt'].notnull()]
dataFinal['Revenue'] = dataFinal['Revenue'].apply(lambda x: abs(x))
```

```python
dataFinal['Last_accr_txn_dt'].max()
```

```python
#Create RFM table
rfmTable = dataFinal.groupby('Customer_ID').agg({'Last_accr_txn_dt': lambda x : (NOW - x.max()).days,
                                                'Invoices': lambda x : sum(x),
                                                'Revenue': lambda x : sum(x)})
rfmTable['Invoices'] = rfmTable['Invoices'].astype(int)
rfmTable.rename(columns={'Last_accr_txn_dt':'recency',
                         'Invoices':'frequency',
                         'Revenue':'Monetary'},
                         inplace=True)
```

```python
rfmTable
```

```python
quantiles = rfmTable.quantile(q=[0.25,0.5,0.75])
quantiles = quantiles.to_dict()
```

```python
segmented_rfm = rfmTable
```

```python
#The lowest recency, highest frequency and monetary amounts are our best customers.
def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]: 
        return 3
    else:
        return 4
    
def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]: 
        return 2
    else:
        return 1
```

```python
segmented_rfm['r_quartile'] = segmented_rfm['recency'].apply(RScore, args=('recency',quantiles,))
segmented_rfm['f_quartile'] = segmented_rfm['frequency'].apply(FMScore, args=('frequency',quantiles,))
segmented_rfm['m_quartile'] = segmented_rfm['Monetary'].apply(FMScore, args=('Monetary',quantiles,))
segmented_rfm.head()
```

```python
segmented_rfm['r_quartile'] = segmented_rfm['r_quartile'].astype(int)
segmented_rfm['f_quartile'] = segmented_rfm['f_quartile'].astype(int)
segmented_rfm['m_quartile'] = segmented_rfm['m_quartile'].astype(int)
```

```python
segmented_rfm['Customer_ID'] = segmented_rfm.index
```

```python
segmented_rfm['RFMScore'] = segmented_rfm.r_quartile.map(str) + segmented_rfm.f_quartile.map(str) + segmented_rfm.m_quartile.map(str)
segmented_rfm.head()
```

```python
rfmWeight = pd.DataFrame()
rfmWeight['r_quartile'] = segmented_rfm['r_quartile']
rfmWeight['f_quartile'] = segmented_rfm['f_quartile']
rfmWeight['m_quartile'] = segmented_rfm['m_quartile']
rfmWeight['RFMScore'] = segmented_rfm['RFMScore']
#rfmWeight['weight']  = rfmWeight.iloc[:,-3:].sum(axis=1)
rfmWeight.reset_index(level=['Customer_ID'], inplace=True)
rfmWeight.head()                        
```

```python
dataFinal = pd.merge(dataFinal,
             rfmWeight[['r_quartile','f_quartile','m_quartile','RFMScore','Customer_ID']],
             on = 'Customer_ID')
```
